Making line plots with matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime,timedelta
```

```python
# generate date list for an year 
start_date = datetime(2024,1,1)
end_date = datetime(2024,12,31)
num_days = (end_date - start_date).days
print(start_date,num_days)
```

```python
date_list =[]
for i in range(num_days):
    date_list.append(start_date.date())
    start_date += timedelta(days=1)

date_list
```

```python
stock_price_list = (np.random.randn(num_days)).tolist()
# stock_price_list
```

```python
np.random.seed(0)
stock_price_list = (np.random.randn(num_days).cumsum()* 50).tolist()
stock_price_list
```

```python
plt.Figure(figsize=(15,6))
plt.plot(date_list,stock_price_list,color ='red',label='price')
plt.title("Stock price for 1 year")
plt.xlabel("date")
plt.ylabel("stock prices")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.Figure(figsize=(15,6))
plt.plot(date_list,stock_price_list,color ='green',marker = '*',label='price')
plt.title("Stock price for 1 year")
plt.xlabel("date")
plt.ylabel("stock prices")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.Figure(figsize=(15,6))
plt.plot(date_list,stock_price_list,color ='green',linestyle = '--',label='price')
plt.title("Stock price for 1 year")
plt.xlabel("date")
plt.ylabel("stock prices")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
stock_price_list_2 = (np.random.randn(num_days).cumsum()* 20 ).tolist()
```

```python
plt.Figure(figsize=(15,6))
plt.plot(date_list,stock_price_list,color ='green',linestyle = '--',label='AAPL')
plt.plot(date_list,stock_price_list_2,color ='red',linestyle = '--',label='MSFT')
plt.title("Stock price for 1 year")
plt.xlabel("date")
plt.ylabel("stock prices")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
